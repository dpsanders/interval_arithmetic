```python
pi = np.pi
pi_half = pi/2.
dospi = 2.*pi
x_list = []
sin_list = []
cos_list = []
for i in range(11):
    q1 = np.random.randint(0,4)
    q2 = np.random.randint(0,4)
    print "ii= ", i, " : ", q1, q2
    if q1 > q2 :
        q2 = q2+4
    
    alow = q1*pi_half + np.random.uniform(0,pi_half)
    ahig = q2*pi_half + np.random.uniform(0,pi_half)
    alow, ahig = min(alow,ahig), max(alow,ahig)
    x_list.append( (alow,ahig) )

    x = Intervalo(alow,ahig)
    ys = sin(x)
    yc = cos(x)
    sin_list.append(ys)
    cos_list.append(yc)
    print " x = ", x
    print "sin( x ) = ", ys
    print "cos( x ) = ", yc
```

```python
ii = 9
alow, ahig = x_list[ii]

xx = np.linspace( alow -1.5*pi, ahig + 1.5*pi, 1000)
plt.plot(xx,np.sin(xx),label="sin(x)")
plt.plot(xx,np.cos(xx),label="cos(x)")
plt.plot(alow,np.sin(alow),'ro')
plt.plot(ahig,np.sin(ahig),'go')
plt.plot(alow,np.cos(alow),'ro')
plt.plot(ahig,np.cos(ahig),'go')
plt.legend()

print alow, ahig
print mp.floor( (alow%dospi)/pi_half ), mp.floor( (ahig%dospi)/pi_half )
print "sin:", sin(Intervalo(alow,ahig) )
print "cos:", cos(Intervalo(alow,ahig) )
```
